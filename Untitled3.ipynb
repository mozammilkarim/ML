```python
import pandas as pd
import seaborn as sns
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
iris.info()
```

```python
iris['species'].value_counts()
# equal distribution of labels
```

using heatmap for correlation

```python
iris.corr()
# will not work for species , since it is pearson'S
# correalation matrix
```

```python
# sepal length is strongly positively correlated to
# petal length and petal width
# so only one of thw above features should be used
# for analysis , otherwise it will add just complexity
```

```python
sns.heatmap(iris.corr())
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='reg',color='g')
# kde is kernel density estimation
```

```python
# records are concentrated at two points
sns.jointplot(x='petal_width',y='petal_length',data=iris,kind='reg',color='b')
sns.jointplot(x='sepal_width',y='petal_length',data=iris,kind='reg',color='g')
```

using paiplots

```python
sns.pairplot(iris,hue='species')
```

```python
# above graphs show clustering of
# species with respect to features
```

using dist plot

```python
sns.distplot(iris['sepal_length'])
sns.distplot(iris['petal_length'])
sns.distplot(iris['sepal_width'])
sns.distplot(iris['petal_width'])
```

now analysis for categorical values

```python
sns.countplot(y=iris['species'])
```

```python
# bar plot
sns.barplot(x='sepal_length',y='species',data=iris)
sns.barplot(x='petal_length',y='species',data=iris)
```

```python
sns.barplot(x='sepal_width',y='species',data=iris)
```

```python
# setosa has widest sepal
# virginica has longest sepal_length and petal_length
```

box plot represents data in percentile data

```python
sns.boxplot(x='petal_length',y='species',data=iris)
```

```python
# there are no outliers in virginica!
```
